```python
#1. Read the csv file into a dataframe called movies_work
movies_work = pd.read_csv("IMDB_horror_movies.csv")
movies_work
```

4. In cast_work, transform the column movie_cast so that it contains, for each movie, a Python list of strings of the actor names (instead of a long string of the actor names separated by “|”). Note that some of the casts are missing so there you should insert an empty list.

```python
#create a dict of all currencies
exchange = {'INR':0.012,'RUR':0.013,'MXN':0.054,'CAD':0.73,'AUD':0.67,'SEK':0.097,'TRL':0.052,
                  'BRL':0.19,'PHP':0.018,'TWD':0.033,'MYR':0.23,'NZD':0.63,'CHF':1.09,'CNY':0.15,
                  'PKR':0.0035,'NOK':0.096,'IDR':0.000066,'DKK':0.15,'SGD':0.75,'HKD':0.13,'ARS':0.0049,
                  'THB':0.029,'NGN':0.0022,'HUF':0.0028,'KRW':0.00078,}
def convertor(x):
    if(pd.isnull(x)):
        return x
    else:
        #remove spaces
        x = x.strip()
        d = float(''.join(c for c in x if (c.isdigit() or c=='.')))
        #if conditions for dollar signs
        if(x[0]=='$'):
            return d
        elif(x[0]=='£'):
            return d*1.23
        elif(x[0]=='€'):
            return d*1.09
        else:
            #use the dictionary for the rest 3 char currency
            return d*exchange[x[:3]]
#131
dollar_budget=dollar_budget.map(convertor)
movies['dollar_budget']=dollar_budget
movies
```

```python
#this is another way using regex, works for plot
def extract_plot(a_list):
    pattern_dir = r"Directed.*?\s\w{2,}\."
    pattern_cast = r"With.*?\s[\w'.]{2,}\."

    director_list = []
    cast_list = []
    plot_list = []
    
    for element in a_list:
        if "Directed" in element:
            director_string = re.search(pattern_dir, element).group()[12:]
            if "With" in element: #has Dir, no With
                cast_string = re.search(pattern_cast, element).group()[5:]
                plot_index = element.find(cast_string)+len(cast_string)+1

            else: #has Dir, no With
                cast_string = ""
                plot_index = element.find(director_string)+len(director_string)+1

        elif "With" in element: #no Dir, has with
            #print(element)
            cast_string = re.search(pattern_cast, element).group()[5:]
            director_string = ""
            plot_index = element.find(cast_string)+len(cast_string)+1

        else: #no Dir, no With
            plot_index = 0
            cast_string = ""
            director_string = ""

        plot_string = element[plot_index:]

        director_list.append(director_string)
        cast_list.append(cast_string)
        plot_list.append(plot_string)
        
    return director_list, cast_list, plot_list
plot = movies['plot'].fillna('')
a,b,c = extract_plot(plot)
c
```
