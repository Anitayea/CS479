4. Create a 13 by 13 array such that the middlemost element is 6. This 6 should be surrounded by all 5's which in turn should be surrounded by 4's, etc. until the edges of the array contain 0's.
Hint: one way to do this is with numpy’s "pad" method.
Also, compute the sum of the numbers in the resulting array.


```python
# start by a 6 in the middle
a = np.ones((1,1))*6
# each time surround the middle with 5 till 0
for i in range(5,-1,-1):
    a = np.pad(a, pad_width=1, constant_values=i)
print(a)
print('sum of the array is: '+str(sum(sum(a))))
```

5. Given the 2D numpy array created with the following list:
[[12, -np.inf, 2, 3], [10, -np.inf, -np.inf, 4], [11, 1, 5, 6]]
Replace the negative infinity values with the average of the positive values in the array.
Thus, the negative infinity values in this case should be replaced by 6 as
(12+2+3+10+4+11+1+5+6)/9 = 6. Calculate this average programmatically (do not just
plug the value 6 in manually).

```python
a = [[12, -np.inf, 2, 3], [10, -np.inf, -np.inf, 4], [11, 1, 5, 6]]
#a function to get mean
def mean(a):
    sum = 0
    len = 0
    for i in a:
        for j in i:
            if(j!= -np.inf):
                sum+=j
                len+=1
    return sum/len
a
```

```python
# replace -np.inf with mean
df=pd.DataFrame([[12, -np.inf, 2, 3], [10, -np.inf, -np.inf, 4], [11, 1, 5, 6]])
df=df.replace(-np.inf, mean(a))
df
```

7. Using pandas, import the file "meals.csv" in our class data folder into a dataframe.
Each row represents the cost of a meal of a particular type (cuisine, e.g., Indian, Italian,
Chinese, etc.)
Given the dataframe, find the five cuisines with the lowest minimum meal prices, the
lowest one first.

```python
meals = pd.read_csv("meals.csv")
price = meals['price']
#slice to get rid of $
meals['price2'] = pd.Series(float(i[1:]) for i in price)
#use groupby to sort and get the min
meals.groupby('cuisine')['price2'].min().sort_values()[:5]
```

8. Again using the meals dataframe, find the total number of meals and the average, minimum, and maximum price of a meal for each cuisine, and the standard deviation of the prices of meals in each cuisine, sorted in ascending order by average price.

```python
count = meals.groupby('cuisine').count()['id']
#get all the central tendencies we need
cp=meals.groupby('cuisine')['price2']
mean = cp.mean()
min = cp.min()
max = cp.max()
std = cp.std()
#put them together in a dataframe
m = pd.DataFrame([count, mean, min, max, std], index = ['count','mean','min','max','std']).T
m.sort_values(by = 'mean')
```

10. Using pandas, import the file "stock_prices.csv" in our class data folder into a dataframe.
Given the dataframe, find the difference between the high and low prices (high minus low) and store this in a new column called "spread". Then sort the dataframe in-place by ticker and then spread and display the modified dataframe.

```python
stock = pd.read_csv('stock_prices.csv')
#add a col spread with the difference between high and low
stock['spread'] = stock['high']-stock['low']
#sort the values based on spread and ticker
stock = stock.sort_values(by=['ticker','spread'])
stock
```
